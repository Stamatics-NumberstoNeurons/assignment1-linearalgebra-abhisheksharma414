## Q3 — Tiny Recommendation System

Each user's taste is represented as a vector of 4 scores across 4 genres — **action, romance, thriller, comedy** — all between 0 and 1.


### Q3 Part (i) — Similarity Matrix

Compute the **full similarity matrix** — every user's similarity score against every other user — using only a **single NumPy operation** (no loops).

What linear algebra concept makes this possible in one line?

````python
# Input — do not modify
users = np.array([
    [0.9, 0.1, 0.8, 0.2],  # User A
    [0.1, 0.9, 0.2, 0.8],  # User B
    [0.8, 0.2, 0.9, 0.1],  # User C
    [0.5, 0.5, 0.5, 0.5],  # User D
])
````

````python
similarity_matrix = users @ users.T
display(similarity_matrix)
````

**What linear algebra concept makes this possible in one line?**


The linear algebra concept that makes computing the full similarity matrix in one line possible is **matrix multiplication (specifically, the dot product)**. When you multiply a matrix by its transpose (`users @ users.T`), each element of the resulting matrix is the dot product of a row vector from the first matrix and a column vector (which was a row vector from the original matrix) from the second matrix. This efficiently computes all pairwise dot products between the user vectors.

### Q3 Part (iii) — The Problem with Raw Dot Products

A senior engineer tells you:
> *"Raw dot products are a bad measure of similarity — you need to normalize."\*

1. What is the **problem** with raw dot products?
2. Construct a **specific example using NumPy** that demonstrates the flaw.
3. Fix it.

> **Hint:** Take a user with scores `[1.8, 0.2, 1.6, 0.4]`

1.  **What is the problem with raw dot products?**
    Raw dot products are sensitive to the **magnitude (length) of the vectors**. A user who gives high scores to many genres (even if their relative preferences are similar to another user) will have a much larger dot product with that other user simply because their vector has a larger magnitude. This can falsely indicate higher similarity between users who merely express stronger preferences overall, rather than having truly aligned tastes. It measures combined intensity rather than similarity in *preference patterns*.

2.  **Construct a specific example using NumPy that demonstrates the flaw.**
    Let's consider `User A = [0.9, 0.1, 0.8, 0.2]` and `User E = [1.8, 0.2, 1.6, 0.4]` (User A's preferences, doubled in intensity). The raw dot product `UserA ⋅ UserE` is `3.0`, while `UserA ⋅ UserA` is `1.5`. This suggests that User A is more similar to User E than to themselves, which is logically flawed. The flaw is that `UserA` and `UserE` have the *exact same preference pattern*, but the raw dot product doesn't reflect this perfect similarity due to the difference in their magnitudes.

3.  **Fix it.**
    The problem can be fixed by **normalizing** the vectors before computing the dot product. This usually means using **cosine similarity**, which divides the dot product by the product of the magnitudes of the vectors. This effectively measures the cosine of the angle between the two vectors, which represents the similarity in their *direction* or *preference pattern*, independent of their magnitude (intensity). After normalization, `cosine_similarity(UserA, UserE)` would correctly be `1.0`, indicating perfect similarity.

````python
import numpy as np

# Define User E based on the hint
user_E = np.array([1.8, 0.2, 1.6, 0.4])

# Original User A
user_A = users[0]
user_B = users[1]

# 1. Demonstrate the flaw with raw dot products
print("--- Raw Dot Product Flaw ---")
print(f"Dot product (User A, User A): {np.dot(user_A, user_A)}")
print(f"Dot product (User A, User E): {np.dot(user_A, user_E)} (User E has same preferences as A, but higher magnitude)")
print(f"Dot product (User A, User B): {np.dot(user_A, user_B)}")
print("\nRaw dot products suggest User A is more similar to E than to itself, which is misleading as E is just A scaled up.")

# 2. Fix it using cosine similarity (normalization)
print("\n--- Normalized Cosine Similarity Fix ---")

def cosine_similarity(vec1, vec2):
    dot_product = np.dot(vec1, vec2)
    norm_vec1 = np.linalg.norm(vec1)
    norm_vec2 = np.linalg.norm(vec2)
    if norm_vec1 == 0 or norm_vec2 == 0:
        return 0.0
    return dot_product / (norm_vec1 * norm_vec2)

print(f"Cosine Similarity (User A, User A): {cosine_similarity(user_A, user_A):.4f}")
print(f"Cosine Similarity (User A, User E): {cosine_similarity(user_A, user_E):.4f}")
print(f"Cosine Similarity (User A, User B): {cosine_similarity(user_A, user_B):.4f}")
print("\nWith cosine similarity, User A and User E correctly show perfect similarity (1.0), as they have the same preference *pattern*.")
````

### Q3 Part (iv) — Most Similar Users

After normalizing, which **two users are most similar**?

Does that match your intuition from just reading the vectors?

What does it mean in plain language — what do these two users have in common?

> **Note:** A similarity score is just a number that tells you **how alike two things are**.

*Write your answer here (double-click to edit this cell)*

After normalizing the vectors and computing cosine similarity:

*   **Which two users are most similar?**
    **User A** and **User C** are the most similar, with a cosine similarity score of approximately **0.9867**. This is very close to 1.0, indicating a strong alignment in their preference patterns.

*   **Does that match your intuition from just reading the vectors?**
    Yes, this matches my intuition. Both User A (`[0.9, 0.1, 0.8, 0.2]`) and User C (`[0.8, 0.2, 0.9, 0.1]`) show high scores for Action and Thriller, and low scores for Romance and Comedy. They both clearly prefer action/thriller genres over romance/comedy. User B, in contrast, prefers romance/comedy, and User D is neutral.

*   **What does it mean in plain language — what do these two users have in common?**
    In plain language, User A and User C have very similar taste profiles. They both strongly enjoy **action** and **thriller** movies, and are not particularly fond of **romance** or **comedy**. They are essentially both 'action/thriller lovers' and would likely enjoy similar movie recommendations.

````python
import numpy as np

# Normalize the user vectors
# Calculate the L2 norm (magnitude) for each user vector
norms = np.linalg.norm(users, axis=1, keepdims=True)
# Divide each user vector by its norm to get unit vectors
normalized_users = users / norms

# Compute the cosine similarity matrix
# This is done by multiplying the normalized_users matrix by its transpose
cosine_similarity_matrix = normalized_users @ normalized_users.T

print("Cosine Similarity Matrix:")
display(cosine_similarity_matrix)

# Find the pair with the highest similarity (excluding self-similarity)
# Fill the diagonal with 0 or -1 to exclude self-similarity
# (or iterate only over upper triangle)
np.fill_diagonal(cosine_similarity_matrix, -1) # Set diagonal to -1 to ignore self-similarity

# Find the indices of the maximum value in the matrix
max_sim_index = np.unravel_index(np.argmax(cosine_similarity_matrix), cosine_similarity_matrix.shape)

# Map indices back to user names
user_labels = ['User A', 'User B', 'User C', 'User D']
most_similar_users = (user_labels[max_sim_index[0]], user_labels[max_sim_index[1]])
max_similarity_score = cosine_similarity_matrix[max_sim_index]

print(f"\nThe two most similar users are {most_similar_users[0]} and {most_similar_users[1]} with a similarity score of {max_similarity_score:.4f}.")
````
